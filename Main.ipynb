```python
import gather
import pickle
import time

batch=1000

with open('userlearn.list', 'rb') as f:
    userlist=pickle.load(f)

with open('iterid.val','rb') as f:
    iterid=pickle.load(f)
    
with open('rating.nda', 'rb') as f:
    R=pickle.load(f)

with open('rated.nda', 'rb') as f:
    I=pickle.load(f)

with open('ratedTest.nda', 'rb') as f:
    It=pickle.load(f)
    
with open('games.list', 'rb') as f:
    gamelist=pickle.load(f)
    
l=10000

while iterid<l:
    print("----------------")
    print("\t",iterid)
    gather.gatherRatingsFromUsers(userlist[:6000],iterid,batch,R,I,It,gamelist)
    
    print("")
    print("#",len(gamelist))
    time.sleep(5)
    iterid+=batch

    with open('iterid.val', 'wb') as f:
        pickle.dump(iterid, f)

    with open('rating.nda', 'wb') as f:
        pickle.dump(R, f)
    
    with open('rated.nda', 'wb') as f:
        pickle.dump(I, f)
        
    with open('ratedTest.nda', 'wb') as f:
        pickle.dump(It, f)
    
    with open('games.list', 'wb') as f:
        pickle.dump(gamelist,f)
```

```python
import pickle
import numpy as np
import blender

with open('cutData.ndas','rb') as f:
    gamelist,userlist,R,I,It = pickle.load(f)

with open('baseline.ndas', 'rb') as f:
    bm,bg,bu = pickle.load(f)

with open('gameinfo.lists','rb') as f:
    namelist, minplist, maxplist, weightlist, catmeclist = pickle.load(f)
    
with open('weightbase.ndas', 'rb') as f:
    cw,bw=pickle.load(f)
    
with open('MF.ndas', 'rb') as f:
    Fg,Fu = pickle.load(f)
   
_TL = np.sum(It)
Iu=np.sum(I,axis=0,keepdims=True)

Rtrivial = np.zeros_like(R) + np.sum(I*R,axis=1,keepdims=True)/np.sum(I,axis=1,keepdims=True)
R0 = bm+np.expand_dims(bg,1)+bu
w = np.expand_dims( np.array(weightlist, dtype="float32"), 1)
wd = np.abs(w-cw)
wsum = np.sum(I*wd, axis=0, keepdims=True)
R1 = R0+bw*(wsum/Iu-wd)
R2= R0+Fg@Fu

ratios=blender.blend(R,I,It,np.stack((Rtrivial,R0,R1,R2)))
print(ratios)
```
